```python
import pandas as pd
import numpy as np

# Set seed for exact reproducibility
np.random.seed(42)
num_rows = 5000

# Generate data features
tx_ids = np.arange(10001, 10001 + num_rows)
dates = pd.date_range(start='2026-01-01', periods=120, freq='D')
random_dates = np.random.choice(dates, size=num_rows)

categories = ['Electronics', 'Clothing', 'Home Appliances', 'Beauty & Health', 'Sports & Outdoors']
regions = ['North', 'East', 'South', 'West']
payment_methods = ['Credit Card', 'PayPal', 'Cash', 'Crypto']

df = pd.DataFrame({
    'Transaction_ID': tx_ids,
    'Date': random_dates,
    'Product_Category': np.random.choice(categories, size=num_rows, p=[0.25, 0.30, 0.20, 0.15, 0.10]),
    'Region': np.random.choice(regions, size=num_rows),
    'Sales_Amount': np.round(np.random.uniform(15.0, 2500.0, size=num_rows), 2),
    'Quantity_Ordered': np.random.randint(1, 11, size=num_rows),
    'Discount_Rate': np.random.choice([0.0, 0.05, 0.10, 0.15, 0.20, 0.25], size=num_rows, p=[0.4, 0.2, 0.15, 0.1, 0.1, 0.05]),
    'Payment_Method': np.random.choice(payment_methods, size=num_rows),
    'Customer_Age': np.random.randint(18, 70, size=num_rows)
})

# Inject missing values and duplicates
df.loc[df.sample(frac=0.03, random_state=12).index, 'Sales_Amount'] = np.nan
df.loc[df.sample(frac=0.02, random_state=34).index, 'Quantity_Ordered'] = np.nan

# Create exact duplicates
duplicates = df.sample(n=50, random_state=56)
df = pd.concat([df, duplicates], ignore_index=True)

# Save to CSV
df.to_csv('company_sales_master.csv', index=False)
print(f"Success! 'company_sales_master.csv' generated with {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
summary_df = df.groupby('Product_Category').agg(Total_Sales=('Sales_Amount', 'sum'),
    Average_Quantity=('Quantity_Ordered', 'mean')).reset_index()

print("summary_df:")
print(summary_df)
```

```python
grouped_df = df.groupby(['Region', 'Product_Category'])['Sales_Amount'].mean().reset_index()
print("grouped_df:")
print(grouped_df)
```
